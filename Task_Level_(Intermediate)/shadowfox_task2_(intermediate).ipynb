**AIR QUALITY INDEX :**

---

The Air Quality Index (AQI) is a measure of how air pollution affects oneâ€™s health within a short time period. It is a tool used by government agencies to communicate to the public how polluted the air is.

The AQI is based on the concentration of pollutants in the air, such as particulate matter (PM10 and PM2.5), ozone, nitrogen dioxide, carbon monoxide, and other pollutants.

**WHAT IS THE PURPOSE OF AIR QUALITY INDEX?**

---



The AQI is designed to help people understand the potential health risks associated with air pollution and take necessary precautions to protect their health. It is an important tool for public health officials, policymakers, and individuals to make informed decisions about outdoor activities and air pollution mitigation strategies.



**There are different levels of Air Quality Index:**

1)Good:0-50:-Air quality is generally safe for everyone.

2)Moderate:51-100:-Air quality is acceptable.Some pollutants may concern for people who are unusually sensitive to air pollution

3)Unhealthy for sensitive groups:101-150: People with respiratory or heart disease, elderly, and childen should take precautions.

4)Unhealthy:151-200: Everyone may begin to experience some adverse health effects.

5)Very unhealthy:201-300: Health alert.Everyone may experience more serious health effects.

6)Hazardous:301-500:Health Warning of emergency conditions.Everyone should avoid all outdoor activities.

**Calculating AQI for each row**

```python
aqi_breakpoints = {
    'co': [0, 4.4, 9.4, 12.4, 15.4, 30.4, 40.4],
    'no2': [0, 53, 100, 360, 649, 1249, 1649],
    'o3': [0, 54, 70, 85, 105, 200, 404],
    'so2': [0, 35, 75, 185, 304, 604, 1004],
    'pm2_5': [0, 12, 35.4, 55.4, 150.4, 250.4, 350.4],
    'pm10': [0, 54, 154, 254, 354, 424, 604],
    'nh3': [0, 53, 104, 154, 204, 304, 404]
}

def calculate_aqi(row):
    max_aqi = 0
    for pollutant in aqi_breakpoints:
        concentration = row[pollutant]
        breakpoints = aqi_breakpoints[pollutant]
        index = next((i for i, x in enumerate(breakpoints) if x >= concentration), len(breakpoints) - 1)
        aqi = ((breakpoints[index] - breakpoints[index - 1]) / (concentration - breakpoints[index - 1])) * (index - 1) + breakpoints[index - 1]
        if aqi > max_aqi:
            max_aqi = aqi
    return max_aqi


df['AQI'] = df.apply(calculate_aqi, axis=1)

df.tail(20)
```

**Categorizing AQI Values**

```python
def categorize_aqi(aqi):
    if 0 <= aqi <= 50:
        return 'Good'
    elif 51 <= aqi <= 100:
        return 'Satisfactory'
    elif 101 <= aqi <= 200:
        return 'Moderately Polluted'
    elif 201 <= aqi <= 300:
        return 'Poor'
    elif 301 <= aqi <= 400:
        return 'Very Poor'
    elif 401 <= aqi <= 500:
        return 'Severe'
    else:
        return 'Invalid AQI Value'


df['Air Quality Category'] = df['AQI'].apply(categorize_aqi)

df.head(20)
```
